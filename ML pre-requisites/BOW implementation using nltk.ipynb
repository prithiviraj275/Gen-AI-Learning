# BOW implementation

```python
import pandas as pd
message = pd.read_csv('ham_spam_1000.csv',names=['label','message'])
```

```python
message
```

```python
import re

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
# sn = SnowballStemmer('english')
```

```python
corpus = []
for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Create Bag of Words

```python
## creating bag of words model
from sklearn.feature_extraction.text import CountVectorizer
## for binary bow enable binary = True
cv = CountVectorizer(max_features=1000,binary=True)
```

```python
x = cv.fit_transform(corpus).toarray()
```

```python
## creating bag of words model with N-grams
from sklearn.feature_extraction.text import CountVectorizer
## for binary bow enable binary = True
cv = CountVectorizer(max_features=1000,binary=True, ngram_range=(1, 2))
x = cv.fit_transform(corpus).toarray()
```
